```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tdc.multi_pred import DTI

davis = DTI(name="davis")
davis.convert_to_log(form = 'binding')

kiba = DTI(name="kiba")
```

```python
sns.set_theme(style="ticks")
plt.figure(figsize=(15, 8))
plt.subplots_adjust(wspace=0.3, hspace=0.2)

plt.subplot(2, 3, 1)
sns.histplot(kiba.get_data().Y, bins=10, binrange=(5, 15))
plt.xlabel("")

plt.subplot(2, 3, 2)
sns.histplot(kiba.get_data().Drug.map(lambda x: len(x)), bins=10, binrange=(0, 150))
plt.xlabel("")
plt.ylabel("")

plt.subplot(2, 3, 3)
sns.histplot(kiba.get_data().Target.map(lambda x: len(x)), bins=20, binrange=(0, 2000))
plt.xlabel("")
plt.ylabel("")

plt.subplot(2, 3, 4)
sns.histplot(davis.get_data().Y, bins=10, binrange=(5, 15))
plt.xlabel("Affinity score")

plt.subplot(2, 3, 5)
sns.histplot(davis.get_data().Drug.map(lambda x: len(x)), bins=10, binrange=(0, 150))
plt.xlabel("Length of Drug sequence")
plt.ylabel("")

plt.subplot(2, 3, 6)
sns.histplot(davis.get_data().Target.map(lambda x: len(x)), bins=20, binrange=(0, 2000))
plt.xlabel("Length of Target sequence")
plt.ylabel("")

plt.show()
```
